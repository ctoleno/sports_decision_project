```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

```

```python
df = pd.read_csv("combined_file_csv_whole.csv")
statcast_raw_df = pd.read_csv("Statcast_2021.csv")
```

```python
df = df.rename(columns = {"Runs_Scored": "runs_scored"})
```

```python
combined = df
bins = list(range(-10, 70, 5))
combined['bucketed_range'] = pd.cut(combined["launch_angle"], bins=bins, right=False)
```

```python
combined["pitch_type"].value_counts()
```

```python

combined_no_runners = combined[(combined["on_1b"] == 0) & ((combined["on_2b"]) == 0) & (combined["on_3b"] == 0)]
combined_1bonly = combined[(combined["on_1b"] == 1) & (combined["on_2b"] == 0) & (combined["on_3b"] == 0)]
combined_2bonly = combined[(combined["on_1b"] == 0) & (combined["on_2b"] == 1) & (combined["on_3b"]== 0)]
combined_3b_only = combined[(combined["on_1b"] == 0) & combined["on_2b"] == 0 & (combined["on_3b"]== 1)]
combined_1stand2nd = combined[(combined["on_1b"] == 1) & (combined["on_2b"] == 1) & (combined["on_3b"] == 0)]
combined_2ndand3rd = combined[(combined["on_1b"] == 0 )& (combined["on_2b"] == 1) & (combined["on_3b"] == 1)]
combined_1stand3rd = combined[(combined["on_1b"] == 1) & (combined["on_2b"] == 0 )& (combined["on_3b"] == 1)]
combined_basesloaded = combined[(combined["on_1b"] == 1) & (combined["on_2b"] == 1) & (combined["on_3b"] == 1)]





def plot_scenario(scenario_df, title):
    fastball = scenario_df[scenario_df["pitch_type"] == "FF"].groupby("bucketed_range")["delta_run_exp"].mean()
    changeup = scenario_df[scenario_df["pitch_type"] == "CH"].groupby("bucketed_range")["delta_run_exp"].mean()
    slider = scenario_df[scenario_df["pitch_type"] == "SL"].groupby("bucketed_range")["delta_run_exp"].mean()

    

    combined_means = pd.DataFrame({
        'FB': fastball,
        'CH': changeup,
        'Slider': slider,
    })

    combined_means.plot(kind='bar', colormap='viridis')

    plt.xlabel('Bucketed Range')
    plt.ylabel('Delta Runs Expected Change')
    plt.title(title)
    plt.show()


plot_scenario(combined_no_runners, 'Delta Runs Expected with No Runners on Base')
plot_scenario(combined_1bonly, 'Delta Runs Expected with Runners on 1st Base Only')
plot_scenario(combined_2bonly, 'Delta Runs Expected with Runners on 2nd Base Only')
plot_scenario(combined_3b_only, 'Delta Runs Expected with Runners on 3rd Base Only')
plot_scenario(combined_1stand2nd, 'Delta Runs Expectedwith Runners on 1st and 2nd Bases')
plot_scenario(combined_2ndand3rd, 'Delta Runs Expected with Runners on 2nd and 3rd Bases')
plot_scenario(combined_2ndand3rd, 'Delta Runs Expected with Runners on 1nd and 3rd Bases')
plot_scenario(combined_basesloaded, 'Delta Runs Expected with Bases Loaded')
```

```python
unique_combinations = combined.groupby(["on_1b", "on_2b", "on_3b", "pitch_type"]).agg(
    mean_delta_run_exp=('delta_run_exp', 'mean'),
    pitch_count=('runs_scored', 'size'),
).reset_index()
unique_combinations = unique_combinations[unique_combinations["pitch_count"] > 500]

# Sort the DataFrame based on mean_runs_scored column in descending order
sorted_combinations = unique_combinations.sort_values(by="mean_delta_run_exp", ascending=False)
sorted_combinations

```

```python
scatter_fig = px.scatter(sorted_combinations, x='pitch_count', y='mean_delta_run_exp', color='pitch_type', hover_data=['on_1b', 'on_2b', 'on_3b'], title='Mean Runs Scored vs. Pitch Count')

# Customize the layout as needed
scatter_fig.update_layout(
    xaxis_title='Pitch Count',
    yaxis_title='Mean Runs Scored',
    legend_title='Pitch Type',
)

# Show the interactive plot
scatter_fig.show()
```

```python
situations_0_outs = unique_combinations[unique_combinations['outs_when_up'] == 0]
situations_1_out = unique_combinations[unique_combinations['outs_when_up'] == 1]
situations_2_outs = unique_combinations[unique_combinations['outs_when_up'] == 2]



# Create scatter plots for each situation
scatter_fig_0_outs = px.scatter(situations_0_outs, x='pitch_count', y='mean_delta_run_exp', color='PlayerType',
                                size='pitch_count', hover_data=["pitch_type",'on_1b', 'on_2b', 'on_3b',"PlayerType"],
                                title='Mean Runs Scored vs. Pitch Count (0 Outs)',
                                labels={'pitch_count': 'Pitch Count', 'mean_delta_run_exp': 'Mean Runs Scored', 'pitch_type': 'Pitch Type'})


scatter_fig_1_out = px.scatter(situations_1_out, x='pitch_count', y='mean_delta_run_exp', color='PlayerType',
                               size='pitch_count', hover_data=["pitch_type",'on_1b', 'on_2b', 'on_3b',"PlayerType"],
                               title='Mean Runs Scored vs. Pitch Count (1 Out)')
scatter_fig_2_outs = px.scatter(situations_2_outs, x='pitch_count', y='mean_delta_run_exp', color='PlayerType',
                                size='pitch_count', hover_data=["pitch_type",'on_1b', 'on_2b', 'on_3b',"PlayerType"],
                                title='Mean Runs Scored vs. Pitch Count (2 Outs)')

# Customize the layout as needed
for fig in [scatter_fig_0_outs, scatter_fig_1_out, scatter_fig_2_outs]:
    fig.update_layout(
        xaxis_title='Pitch Count',
        yaxis_title='Mean Runs Scored',
        legend_title='Pitch Type',
    )

# Show the interactive plots
scatter_fig_0_outs.show()
scatter_fig_1_out.show()
scatter_fig_2_outs.show()
```
